```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
%matplotlib inline
```

```python
V = 5
T = 1
N = 500
dt = T/N

t = np.linspace(0, T, N, endpoint=False)

plt.plot(t, V-V*sig.square(2 * np.pi * t))
plt.grid()
```

```python
def ComputeBn(V,n_array):
    
    bn_array = (2*V/(n_array*np.pi))*(np.cos(n_array*np.pi)-1)
    
    return bn_array

def ComputeV(V,T,n,t):
    
    V_array = np.ones(len(t))*V

    for i in range(0,n):
    
        V_add = bns[i]*np.sin(2*np.pi*(i+1)/T*t)
        V_array += V_add
        
    return V_array

n = 40
n_array = np.arange(1, n+1)   
bns = ComputeBn(V,n_array)
V_array = ComputeV(V,T,n,t)

plt.plot(t, V-V*sig.square(2 * np.pi * t))
plt.plot(t,V_array)
plt.grid()
```
